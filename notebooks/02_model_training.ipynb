```python
import numpy as np      
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df= pd.read_csv('C:/Users/memer/OneDrive/Documentos/Eme/UNIVERSIDAD/2025-1/ETL/Tercer Cohorte/Proy85/proy85/data/dfobjective.csv')
```

```python
df = pd.get_dummies(df, columns=['listing_type_id'], drop_first=True)
```

```python
X = df.drop(columns=['condition'])  
y = df['condition']                 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model1 = RandomForestClassifier(random_state=42)
model1.fit(X_train_scaled, y_train)
```

```python
y_pred1 = model1.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
```

```python
model2 = LogisticRegression()
model2.fit(X_train_scaled, y_train)
```

```python
y_pred2 = model2.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred2)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred2))
```

```python
model = SVC(kernel='linear') 
model.fit(X_train_scaled, y_train)
y_pred3 = model.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred3)
print(f'Accuracy: {accuracy}')
report = classification_report(y_test, y_pred3)
print('Classification Report:')
print(report)
```

```python
model4 = KNeighborsClassifier(n_neighbors=5) 
model4.fit(X_train_scaled, y_train)
y_pred4 = model4.predict(X_test_scaled)
```

```python
accuracy4 = accuracy_score(y_test, y_pred4)
print(f'Accuracy for KNN: {accuracy4}')

cm4 = confusion_matrix(y_test, y_pred4)
print('Confusion Matrix for KNN:')
print(cm4)

report4 = classification_report(y_test, y_pred4)
print('Classification Report for KNN:')
print(report4)
```

```python

model5 = DecisionTreeClassifier(random_state=42)
model5.fit(X_train_scaled, y_train)
y_pred5 = model5.predict(X_test_scaled)

accuracy5 = accuracy_score(y_test, y_pred5)
print(f'Accuracy for Decision Tree: {accuracy5}')
print(confusion_matrix(y_test, y_pred5))
print(classification_report(y_test, y_pred5))
```
